<h1> Experiment 1 </h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.integrate import quad


n= 300
h = 0.01
a = np.zeros(n)
Inum = np.zeros(n)
Iden = np.zeros(n)
errnum = np.zeros(n)
errden = np.zeros(n)
Exp_val = np.zeros(n)


def fnum(x,a):
    return np.exp(-2*a*x**2)*(2*a+ x**2*(1-4*a**2))

def fden(x,a):
    return 2*np.exp(-2*a*x**2)


for i in range(0,n):
    a[i] = (i*h)
    Inum[i], errnum[i] = quad(fnum, 0 , np.inf ,args=(a[i],))
    Iden[i], errden[i] = quad(fden, 0 , np.inf ,args=(a[i],))
    Exp_val[i] = Inum[i]/Iden[i]


#Finding the Minimum Value 
Exp_val = Exp_val[1:]
a = a[1:]
min_val = np.argmin(Exp_val)
print("Ground State Energy: ", a[min_val])



plt.plot(a[1:n],Exp_val[1:n])
plt.xlabel('Alpha')
plt.xticks(np.arange(0, 3.5, 0.5))
plt.ylabel('Expectation Value')
plt.title('Expectation value vs Alpha plot')
plt.show()
```
