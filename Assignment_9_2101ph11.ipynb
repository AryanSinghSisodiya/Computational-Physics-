```python
import numpy as np 
import matplotlib.pyplot as plt

def f(x):
    return x**2

t = np.arange(0, 1000, 0.1)

negative_values = [-0.5, -0.3,-0.6,0]
positive_values =[ 0.3, 0.5, 0.6]
for u0 in negative_values:
    u = [u0]
    h = 0.001
    
    for i in range(3):
        var = u[-1] + h * f(u[-1])
        u.append(var)

    for i in range(4,10000):
        var = u[-1] + h * (55*f(u[-1]) - 59*f(u[-2]) + 37*f(u[-3]) - 9*f(u[-4]))/24
        u.append(var)

    plt.plot(t, u, label=f'u[0]={u0}')

for u0 in positive_values:
    u = [u0]
    h = 0.0001
    
    for i in range(3):
        var = u[-1] + h * f(u[-1])
        u.append(var)

    for i in range(4,10000):
        var = u[-1] + h * (55*f(u[-1]) - 59*f(u[-2]) + 37*f(u[-3]) - 9*f(u[-4]))/24
        u.append(var)

    plt.plot(t, u, label=f'u[0]={u0}')

plt.xlabel('t')
plt.ylabel('u')
plt.legend()
plt.show()
```
