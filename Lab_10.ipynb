```python
import numpy as np 
import matplotlib.pyplot as plt
import random

def f(x):
    return x**2

t = np.arange(0, 1000, 0.1)
positive_values = np.zeros(10)
negative_values = np.zeros(10)



for i in range(10):
    negative_values[i] = random.uniform(-2,0)
    positive_values[i] = random.uniform(0,2)

print(negative_values)
for u0 in negative_values:
    u = [u0]
    h = 0.0001
    
    for i in range(3):
        var = u[-1] + h * f(u[-1])
        u.append(var)

    for i in range(4,10000):
        var = u[-1] + h * (55*f(u[-1]) - 59*f(u[-2]) + 37*f(u[-3]) - 9*f(u[-4]))/24
        u.append(var)

    plt.plot(t, u)

for u0 in positive_values:
    u = [u0]
    h = 0.00005
    
    for i in range(3):
        var = u[-1] + h * f(u[-1])
        u.append(var)

    for i in range(4,10000):
        var = u[-1] + h * (55*f(u[-1]) - 59*f(u[-2]) + 37*f(u[-3]) - 9*f(u[-4]))/24
        u.append(var)

    plt.plot(t, u)

plt.xlabel('t')
plt.ylabel('u')
plt.legend()
plt.show()
```
