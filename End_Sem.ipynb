```python
import numpy as np
import matplotlib.pyplot as plt


m = 1  
k = 6  
x0 = 0.5  
t_max = 6  
dt = 0.01  

def force(x):
    return -k*x**3

t = np.arange(0, t_max, dt)
x = np.zeros_like(t)
v = np.zeros_like(t)

x[0] = x0

for i in range(1, len(t)):
    v_half = v[i-1] + 0.5 * dt * force(x[i-1]) / m
    x[i] = x[i-1] + dt * v_half
    v[i] = v_half + 0.5 * dt * force(x[i]) / m

plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Position of the Particle')

plt.subplot(2, 1, 2)
plt.plot(t, v)
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity of the Particle')

plt.tight_layout()
plt.show()
```
